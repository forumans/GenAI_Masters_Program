# FitPro – AI-Powered Digital Marketing Plan

**Domain:** Digital Marketing  
**Technology:** ChatGPT (OpenAI API)

## Problem Statement
FitPro, a leading e-commerce retailer of fitness equipment, needs to enhance its online presence and digital marketing strategy. This notebook uses ChatGPT to build a comprehensive digital marketing plan covering SEO, content marketing, email campaigns, FAQs, and a content calendar.

## Deliverables
1. Top 20 SEO Keywords related to fitness
2. Blog post built around the SEO keywords
3. Email campaign plan
4. Sample marketing emails
5. Top 10 customer FAQs for fitness
6. Effective content calendar

---
## Setup – Import Libraries & Configure OpenAI

```python
import os
from openai import OpenAI

# ----------------------------------------------------------------
# Set your OpenAI API key in a .env file as:
#   OPENAI_API_KEY=sk-...
# ----------------------------------------------------------------
# Load from .env file (recommended)

from dotenv import load_dotenv
load_dotenv()  # loads OPENAI_API_KEY from .env

client = OpenAI(api_key=os.environ.get('OPENAI_API_KEY'))

# Helper – single call wrapper
def ask_chatgpt(prompt, model='gpt-3.5-turbo', temperature=0.7, max_tokens=1500):
    """Send a prompt to ChatGPT and return the response text."""
    response = client.chat.completions.create(
        model=model,
        messages=[
            {
                'role': 'system',
                'content': (
                    'You are an expert digital marketing specialist with deep knowledge '
                    'in SEO, content marketing, email marketing, and social media strategy '
                    'for the fitness and wellness industry.'
                )
            },
            {'role': 'user', 'content': prompt}
        ],
        temperature=temperature,
        max_tokens=max_tokens
    )
    return response.choices[0].message.content

print('OpenAI client configured. Ready to generate content.')
```

---
## 1. Top 20 SEO Keywords Related to Fitness

```python
seo_prompt = """
You are an SEO expert for FitPro, an e-commerce retailer selling fitness equipment online.

Identify the top 20 SEO keywords related to fitness and fitness equipment that FitPro 
should target to increase organic search traffic and drive online sales.

For each keyword provide:
- The keyword phrase
- Search intent (informational / transactional / navigational)
- Estimated competition level (Low / Medium / High)
- Why it is valuable for FitPro

Present the results as a numbered list.
"""

seo_keywords = ask_chatgpt(seo_prompt, max_tokens=2000)
print('=== TOP 20 SEO KEYWORDS FOR FITPRO ===')
print(seo_keywords)
```

---
## 2. Blog Post Using SEO Keywords

```python
blog_prompt = """
Write a comprehensive, SEO-optimised blog post for FitPro's website.

Requirements:
- Title: Make it compelling and keyword-rich
- Word count: approximately 600–800 words
- Naturally incorporate at least 10 of these fitness SEO keywords:
  home gym equipment, fitness equipment, workout gear, resistance bands,
  adjustable dumbbells, treadmill for home, best exercise bike, weight bench,
  cardio equipment, strength training equipment, gym equipment online,
  affordable fitness equipment, buy dumbbells online, yoga mat, pull-up bar
- Structure: Introduction, 3–4 subheadings (H2), Conclusion with a CTA
- Tone: Motivational, friendly, and informative
- Include a meta description (max 160 characters) at the top
"""

blog_post = ask_chatgpt(blog_prompt, max_tokens=2000)
print('=== FITPRO SEO BLOG POST ===')
print(blog_post)
```

---
## 3. Email Campaign Plan

```python
campaign_prompt = """
Design a 30-day email marketing campaign plan for FitPro (fitness equipment e-commerce).

The campaign goal is to:
- Nurture new subscribers and convert them into buyers
- Re-engage inactive customers
- Promote seasonal offers and new product launches

Provide a structured plan with:
1. Campaign name and overall theme
2. Target audience segments (at least 3)
3. Email sequence (Week 1 to Week 4) – for each email specify:
   - Day / timing
   - Email type (Welcome / Educational / Promotional / Re-engagement)
   - Subject line
   - Key message / goal
4. Key performance metrics to track
"""

campaign_plan = ask_chatgpt(campaign_prompt, max_tokens=2000)
print('=== FITPRO EMAIL CAMPAIGN PLAN ===')
print(campaign_plan)
```

---
## 4. Sample Marketing Emails

```python
# Email 1 – Welcome Email
welcome_prompt = """
Write a warm and engaging Welcome Email for FitPro's new subscribers.

Details:
- Brand: FitPro (online fitness equipment retailer)
- Goal: Make the subscriber feel valued, introduce the brand, and drive first purchase
- Include: Subject line, preview text, greeting, brand story (2–3 sentences),
  what they can expect from FitPro, a 10% discount offer for first purchase,
  a clear CTA button text, and a friendly sign-off
- Tone: Warm, motivating, and professional
"""

welcome_email = ask_chatgpt(welcome_prompt, max_tokens=800)
print('=== EMAIL 1: WELCOME EMAIL ===')
print(welcome_email)
```

```python
# Email 2 – Promotional / Sales Email
promo_prompt = """
Write a high-converting promotional email for FitPro's Summer Fitness Sale.

Details:
- Sale: Up to 40% off on home gym equipment
- Duration: Sale ends in 3 days (create urgency)
- Featured products: Adjustable Dumbbells, Resistance Band Set, Foldable Treadmill
- Include: Subject line, preview text, headline, offer details, product highlights
  (3 products with short descriptions and sale prices), urgency statement,
  CTA button text, and footer note
- Tone: Exciting, urgent, action-driven
"""

promo_email = ask_chatgpt(promo_prompt, max_tokens=800)
print('=== EMAIL 2: PROMOTIONAL / SALES EMAIL ===')
print(promo_email)
```

```python
# Email 3 – Re-engagement Email
reengagement_prompt = """
Write a re-engagement email for FitPro targeting customers who haven't purchased 
in the last 90 days.

Details:
- Goal: Win back inactive customers and remind them of FitPro's value
- Include: Subject line, preview text, empathetic opening (acknowledge the gap),
  what's new at FitPro (2–3 updates), a special 'We Miss You' discount (15% off),
  CTA button text, and an unsubscribe-friendly footer
- Tone: Friendly, empathetic, and low-pressure
"""

reengagement_email = ask_chatgpt(reengagement_prompt, max_tokens=800)
print('=== EMAIL 3: RE-ENGAGEMENT EMAIL ===')
print(reengagement_email)
```

```python
# Email 4 – Educational / Value Email
educational_prompt = """
Write an educational email for FitPro subscribers focused on providing fitness value.

Details:
- Topic: '5 Must-Have Home Gym Equipment Pieces for Beginners'
- Goal: Build trust and authority, subtly promote FitPro products
- Include: Subject line, preview text, intro paragraph, list of 5 equipment items
  (each with a 2-sentence description of the benefit), a soft CTA to shop at FitPro,
  and a motivational sign-off
- Tone: Helpful, informative, and encouraging
"""

educational_email = ask_chatgpt(educational_prompt, max_tokens=1000)
print('=== EMAIL 4: EDUCATIONAL / VALUE EMAIL ===')
print(educational_email)
```

## 5. Top 10 Customer FAQs for Fitness

```python
faq_prompt = """
You are a customer service expert for FitPro, an online fitness equipment retailer.

List the top 10 most frequently asked questions (FAQs) customers ask related to:
- Fitness equipment selection and usage
- Home gym setup
- Orders, shipping, and returns (FitPro-specific)
- Fitness goals and workout tips

For each FAQ provide:
- The question (as a customer would ask it)
- A clear, helpful, and friendly answer (3–5 sentences)

Format as a numbered list with Q: and A: labels.
"""

faqs = ask_chatgpt(faq_prompt, max_tokens=2000)
print('=== TOP 10 CUSTOMER FAQs – FITPRO ===')
print(faqs)
```

## 6. Content Calendar

```python
calendar_prompt = """
Create a comprehensive 4-week content calendar for FitPro's digital marketing across 
multiple channels.

Channels to cover:
- Blog / Website (2 posts/week)
- Instagram (5 posts/week)
- Facebook (3 posts/week)
- Email Newsletter (1 email/week)
- YouTube / Video (1 video/week)

For each content piece include:
- Day and channel
- Content type (Reel / Carousel / Blog / Email / Video / Story)
- Topic / Title
- Goal (Awareness / Engagement / Conversion / Retention)
- Key message or hook (1 sentence)

Present as a structured weekly table (Week 1, Week 2, Week 3, Week 4).
Align content themes with fitness motivation: 
Week 1 = New Year New You, Week 2 = Build Your Home Gym, 
Week 3 = Expert Tips & Education, Week 4 = Sale & Conversion.
"""

content_calendar = ask_chatgpt(calendar_prompt, max_tokens=2500)
print('=== FITPRO 4-WEEK CONTENT CALENDAR ===')
print(content_calendar)
```
